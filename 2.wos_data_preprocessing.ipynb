```python
import pandas as pd
import glob
import os
import json


```

```python
def read_and_combine_excel(folder_path, file_pattern="*.xls"):
    # Initialize an empty list to hold DataFrames
    all_dataframes = []
    
    # Create a file list based on the folder_path and file_pattern
    file_list = glob.glob(os.path.join(folder_path, file_pattern))
    
    # Iterate over each file, read it into a DataFrame, and append to the list
    for file in file_list:
        # Read the Excel file
        df = pd.read_excel(file)  # dtype=str ensures all data is string; adjust as needed.
        all_dataframes.append(df)
    
    # Combine all DataFrames into one
    if all_dataframes:
        combined_df = pd.concat(all_dataframes, ignore_index=True)
    else:
        # If no files found, return an empty DataFrame
        combined_df = pd.DataFrame()
    
    return combined_df

def dataframe_to_json(df, output_path="combined_data.json"):
    # Convert DataFrame rows to a list of dictionaries
    data_list = df.to_dict(orient='records')
    
    # Write the list of dictionaries to a JSON file
    with open(output_path, 'w', encoding='utf-8') as json_file:
        json.dump(data_list, json_file, ensure_ascii=False, indent=4)

```

```python
folder_path = r"/home/chris/Data/Projects/playground/articles"

# Read and combine all Excel files in the given folder
combined_df = read_and_combine_excel(folder_path)

# Convert the combined DataFrame into JSON and save it
dataframe_to_json(combined_df, "unique_entries.json")

print("Data successfully combined and written to unique_entries.json")
```
